```python
def header(titulo, largura=50):
    """Nível 1: Títulos de seções principais."""
    print("\n" + "=" * largura)
    print(f"{titulo.upper():^{largura}}")
    print("-" * largura)

def status(msg):
    """Nível 2: Ações ou estados atuais do código."""
    print(f"\n  > {msg}...")

def param(nome, valor, unidade=""):
    """Nível 3: Valores de variáveis e constantes."""
    # Formata a unidade se ela existir
    unit_str = f" [{unidade}]" if unidade else ""
    # Exibe com um recuo maior para mostrar subordinação
    print(f"      {nome} = {valor}{unit_str}")

# --- Exemplo de uso aplicado ao seu workflow de Astrofísica ---

header("Configuração do Modelo JPO")
status("Carregando parâmetros estelares")
param("M_star", 0.08, "M_sun")
param("R_star", 0.117, "R_sun")
param("T_eff", 2550, "K")

status("Parâmetros do Campo Magnético")
param("B_0", 600, "G")
param("f_0", 0.1, "") # Sem unidade

header("Iniciando Integração")
status("Buscando ponto crítico via Shooting Method")
param("u_0_guess", 10.5, "km/s")
param("tol", "1e-6", "")

# ... lógica do código ...

status("Resultado obtido")
param("u_critico", 15.42, "km/s")
print("-" * 50)
```

```python
def header(titulo, largura=60, **kwargs):
    """
    Nível 1: Títulos de seções principais.
    Aceita parâmetros extras para exibir uma 'configuração de base'.
    """
    print("\n" + "=" * largura)
    print(f"{titulo.upper():^{largura}}")
    
    # Se houver parâmetros extras (a=2, b=4), eles aparecem aqui
    if kwargs:
        # Cria uma string do tipo "a: 2  |  b: 4"
        info_str = "  |  ".join([f"{k}: {v}" for k, v in kwargs.items()])
        print(f"{info_str:^{largura}}")
        
    print("-" * largura)

def status(msg):
    """Nível 2: Ações ou estados atuais do código."""
    print(f"\n  > {msg}...")

def param(nome, valor, unidade=""):
    """Nível 3: Valores de variáveis e constantes detalhadas."""
    unit_str = f" [{unidade}]" if unidade else ""
    print(f"      {nome} = {valor}{unit_str}")

# --- Exemplo de uso Prático ---

# Aqui você passa os parâmetros de base diretamente no header
header("Iniciando Simulação MHD", a=2, b=4, modelo="JPO")

status("Processando variáveis iniciais")
param("u_0", 15.5, "km/s")
param("n_e", "10^6", "cm^-3")

status("Integração numérica")
param("passo", 0.001, "")
param("iteração", 450, "")

# Se quiser um header simples sem parâmetros, ele continua funcionando:
header("Resultados Finais")
param("u_final", 300.2, "km/s")
print("-" * 60)
```
